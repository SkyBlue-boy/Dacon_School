### 문제 1. data_mdc02 폴더에 있는 train.csv와 test.csv를 각각 읽어서 train, test 변수에 저장하세요.

```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv', index_col=0) # index_col=0은 첫번째 열값을 행의 레이블로 
```

### train 데이터의 type을 sample_submission에 대응하는 가변수 형태로 변환시킵니다.

```python
column_number = {}
for i, column in enumerate(sample_submission.columns):
    column_number[column] = i
    
def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

### 문제 2. train, test 변수의 상위 5개 행을 확인해보세요.

```python
print(train.head(5))
print(test.head(5))
```

각 컬럼에 대한 설명은 아래와 같습니다.

> type: 천체 유형 \
> fiberID: 관측에 사용된 광섬유의 구분자 \
> psfMag: 먼 천체를 한 점으로 가정하여 측정한 빛의 밝기 (Point spread function magnitudes)\
> fiberMag: 3인치 지름의 광섬유를 사용하여 광스펙트럼을 측정할 때, 광섬유를 통과하는 빛의 밝기 (Fiber magnitudes)\
> petroMag: 은하처럼 뚜렷한 표면이 없는 천체에서는 빛의 밝기 측정이 제한됨. 천체의 위치와 거리에 상관없이 빛의 밝기를 비교하기 위한 수치 (Petrosian magnitudes)\
> modelMag: 천체 중심으로부터 특정 거리의 밝기 (Model magnitudes)\

### 문제 4. train 데이터에서 psfMag_u, fiberMag_g, petroMag_r, modelMag_i 칼럼만 추출하여 X_train 변수에 저장하고, type 칼럼을 추출하여 y_train 변수에 저장해보세요.

```python
X_train = train[['psfMag_u', 'fiberMag_g', 'petroMag_r', 'modelMag_i']]
X_train
```

```python
y_train = train[['type_num']] # 위에서 train 데이터의 type을 sample_submission에 대응하는 가변수 형태로 변환했기 때문에 type칼럼 대신 type_num 칼럼을 추출
y_train.head()
```

### 문제 5. RandomForestClassifer을 불러와서 random_state는 7로 고정하여 model 변수에 저장해보세요.

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 7, verbose = 1, n_jobs = -1)
#verbose : 실행 과정 출력 여부
#n_jobs :적합성과 예측성을 위해 병렬로 실행할 작업 수(일꾼들), n_jobs=-1로 지정하면 컴퓨터의 모든 일꾼들을 사용
#random_state : 난수 seed 설정, 랜덤하게 만들어지기 때문에 random_state를 고정해야 어떤 컴퓨터에서도 같은 결과를 볼 수 있다. 
```

### 문제 6. 문제 4에서 정의한 X_train, y_train을 활용하여 model을 학습시켜 보세요.

```python
model.fit(X_train, y_train)
```

### 문제 7. test 데이터에 대한 예측값을 산출해 주세요.

```python
X_test = test[['psfMag_u', 'fiberMag_g', 'petroMag_r', 'modelMag_i']]
```

```python
sorted(test.unique())
```

### 점수 높이기
4가지 feature만 가지고 train을 했기 때문에 점수가 낮다고 판단해서 모든 feature를 가지고 train을 하기로 함

```python
X_train2 = train.drop(columns=['type', 'type_num'], axis=1) #axis=1은 열을 기준으로 drop하기 위해
y_train2 = train[['type_num']]
X_test2 = test
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 7, verbose =1, n_jobs = -1)
```

```python
model.fit(X_train2, y_train2)
```

```python
pred2 = model.predict_proba(X_test2)
```

```python
submission = pd.DataFrame(data=pred2, columns = sample_submission.columns, index=sample_submission.index)
submission.to_csv('featall.csv', index=True) #index=True -> 행의 이름 index 도 같이 내보냄.
```
